## Uji Frekuensi / Uji Normalitas

`Kolmogorov Smirnoff (KS)`

$H_{0}$ : Data isn't Normally Distributed

$H_{a}$ : Data is Normally Distributed

- If the P-value of the KS Test larger than 0.05 (Reject H_0, we assume a normal distribution)
- If the P-value of the KS Test smaller than 0.05 (Failed to reject H_0, we assume not a normal distribution)

```python
from scipy.stats import kstest

kstest(df['total_bill'],'norm')
```

p-value = 0.00 < 0.05

Karena p-value < 0.05, maka menurut uji KS, kolom total bill tidak terdistribusi normal

`Liliefors`

$H_{0}$ : Data isn't Normally Distributed

$H_{a}$ : Data is Normally Distributed

- If the P-value of the Liliefors Test larger than 0.05 (Reject H_0, we assume a normal distribution)
- If the P-value of the Liliefors Test smaller than 0.05 (Failed to reject H_0, we assume not a normal distribution)

```python
from statsmodels.stats.diagnostic import lilliefors

lilliefors(df['total_bill'])
```

p-value = 0.0009999999999998899 < 0.05

Karena p-value < 0.05, maka menurut uji Lilliefors, kolom total bill tidak terdistribusi normal

`Shapiro Wilk`

$H_{0}$ : Data isn't Normally Distributed

$H_{a}$ : Data is Normally Distributed

- If the P-value of the Saphiro Wilk Test larger than 0.05 (Reject H_0, we assume a normal distribution)
- If the P-value of the Saphiro Wilk Test smaller than 0.05 (Failed to reject H_0, we assume not a normal distribution)

```python
from scipy.stats import shapiro

shapiro(df['total_bill'])
```

p-value = 3.3245391868090786e-10 < 0.05

Karena p-value < 0.05, maka menurut uji Shapiro, kolom total bill tidak terdistribusi normal

`Normaltest`

$H_{0}$ : Data isn't Normally Distributed

$H_{a}$ : Data is Normally Distributed

- If the P-value of the Normaltest Test larger than 0.05 (Reject H_0, we assume a normal distribution)
- If the P-value of the Normaltest Test smaller than 0.05 (Failed to reject H_0, we assume not a normal distribution)

```python
from scipy.stats import normaltest

normaltest(df['total_bill'])
```

p-value = 1.5951078766352608e-10 < 0.05

Karena p-value < 0.05, maka menurut uji Normaltest, kolom total bill tidak terdistribusi normal

```python
from scipy.stats import normaltest
p_val=normaltest(df['total_bill'])[1]
if p_val<=0.05:
    print(f"Karena p-value ≤ 0.05, maka menurut uji Normaltest, kolom total bill tidak terdistribusi normal")
else:
    print(f"Karena p-value > 0.05, maka menurut uji Normaltest, kolom total bill terdistribusi normal")

```
