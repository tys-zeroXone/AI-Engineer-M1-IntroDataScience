## Missing Values

Nilai kosong dalam dataframe

```python
df.isna()
```

```python
df.isna().sum()
```

```python
# menampilkan baris yg OverTimePay-nya missing
df[df['OvertimePay'].isna()]
```

```python
index_missing = df[df['OvertimePay'].isna()].index
index_missing 
```

```python
# mau mengisi data kosong dengan rata-rata

# cari mean dari OvertimePay 
df['OvertimePay'].mean()
```

```python
# mengisi data kosong di kolom OvertimePay dengan rata-rata OvertimePay keseluruhan
df['OvertimePay'].fillna(value= df['OvertimePay'].mean()) 
```

```python
# simpan dalam series (df['OvertimePay']), bukan dalam dataframe (df)
df['OvertimePay'] = df['OvertimePay'].fillna(value= df['OvertimePay'].mean()) 
```

```python
# yg tadinya missing, sekarang sudah terisi dengan mean-nya
df['OvertimePay'].loc[index_missing] 
```

```python
df.isna().sum()
```

```python
# mengisi data yg missing pada suatu kolom dengan string
df['Notes'] = df['Notes'].fillna(value='Unknown')
df.head()

# kalau kolomnya kategorikal (string), 
# maka bisa diisi dengan mode (nilai yg paling sering muncul) atau dengan string baru ('other' atau 'unknown')

# kalau kolomnya numerikal (int atau float), maka bisa diisi dengan mean atau median
```

```python
df.isna().sum()
```

```python
# menghapus kolom yg isinya sangat banyak yg missing
df = df.drop(columns='Status')
df.head()
```

```python
df.isna().sum()
```

```python
# jika salah satu data pada suatu baris ada yg missing (NaN), maka baris tsb akan dihapus
df.dropna()
```

```python
# drop baris yg BasePay-nya missing
df.dropna(subset='BasePay')
```

## Data Aggregation

Proses pengumpulan atau penyusunan data menjadi sebuah rangkuman

```python
data = {
    "Company": ["GOOG", "GOOG", "MSFT", "MSFT", "FB", "FB"],
    "Person": ["Sam", "Charlie", "Amy", "Vanessa", "Carl", "Sarah"],
    "Sales": [200, 120, 340, 124, 243, 350]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby('Company')
```

```python
by_comp = df.groupby('Company')[['Sales']]
```

```python
by_comp.mean()
```

```python
df.groupby('Company')[['Sales']].mean()
```

```python
by_comp.std()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()["GOOG"]
```

```python
by_comp.describe().transpose()["GOOG"].loc["Sales"].loc["25%"]
```
