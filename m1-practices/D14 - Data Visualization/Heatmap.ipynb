```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import cufflinks as cf
cf.go_offline()

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

```python
df=sns.load_dataset('tips')
df
```

```python
#matriks
df_num=df.select_dtypes(include='number').corr()
df_num
```

# Heatmap

## Matplotlib

```python
plt.figure(figsize=(8,8))
plt.pcolor(df_num,cmap='Blues')
plt.title('Heatmap Data Tips')
```

## Seaborn

```python
# plt.figure(figsize=(6,6))
sns.heatmap(df_num,cmap='Blues',annot=True,fmt='.2g')
plt.title('Heatmap Data Tips')
plt.show()

```

```python

sns.pairplot(df)
plt.show()
```

```python
corr=df_num.corr()
matrix = np.triu(corr)

# plt.figure(figsize=(6,6))
sns.heatmap(df_num,cmap='Blues',annot=True,fmt='.2g',mask=matrix)
plt.title('Heatmap Data Tips')
plt.show()
```

```python
for i in sns.get_dataset_names():
    if i!='brain_networks':
        print(i)
        df=sns.load_dataset(i).select_dtypes(include='number').corr()
        df=df.dropna()
        sns.heatmap(df,annot=True,cmap='Blues')
        plt.show()
    
```
