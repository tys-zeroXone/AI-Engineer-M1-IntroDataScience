### Scatterplot 3d

```python
df=sns.load_dataset('tips')

## scatter 3D : Tips (Total Bill, Size, Tip)
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection='3d')
ax.scatter(data=df,xs='total_bill',ys='size',zs='tip')
ax.set_xlabel('Total Bill ($)')
ax.set_ylabel('Size')
ax.set_zlabel('Tip ($)')
plt.title('Total Bill vs Size vs Tip')
plt.show()
```

```python
ax=sns.barplot(data=tips_by_day,x='day',y='tip')
plt.title('Average of Tip per Day')
plt.xlabel('Day')
plt.ylabel('Average of Tip ($)')
plt.grid()

# Menambahkan anotasi pada setiap batang plot
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()-0.07), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')

plt.show()
```

```python
count_by_day=tips.groupby('day')['tip'].count().reset_index().rename(columns={'tip':'count'})
# Menambahkan anotasi dalam bentuk persentase pada setiap batang plot
# Menghitung total jumlah data
count_by_day['total_count'] = count_by_day['count'].sum()

# Menghitung persentase
count_by_day['percentage'] = (count_by_day['count'] / count_by_day['total_count']) * 100

count_by_day
```

```python
ax=sns.barplot(data=count_by_day,x='day',y='count')
plt.title('Count of Customer per Day')
plt.xlabel('Day')
plt.ylabel('Count of Customer')
plt.grid()

# Menambahkan anotasi pada setiap batang plot
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()-2), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')

plt.show()
```

```python
ax=sns.barplot(data=count_by_day,x='day',y='percentage')
plt.title('Count of Customer per Day')
plt.xlabel('Day')
plt.ylabel('Percentage')
plt.grid()


for p in ax.patches:
    percentage = '{:.1f}%'.format( p.get_height() )
    ax.annotate(percentage, 
                   (p.get_x() + p.get_width() / 2., p.get_height()-0.8), 
                   ha='center', va='center', 
                   xytext=(0, 10), 
                   textcoords='offset points')
plt.show()
```

```python
ax=sns.barplot(data=count_by_day,x='count',y='day')
plt.title('Count of Customer per Day')
plt.xlabel('Day')
plt.ylabel('Count of Customer')
plt.grid()

# Menambahkan anotasi pada setiap batang plot
for p in ax.patches:
    ax.annotate(format(p.get_width(), '.0f'), 
                (p.get_width(), p.get_y() + p.get_height() / 2),  # posisi teks
                ha='left', va='center', 
                xytext=(5, 0), 
                textcoords='offset points') 

plt.show()

```

```python
plt.figure(figsize=(15,8))
ax=sns.barplot(data=count_by_day,x='percentage',y='day')
plt.title('Count of Customer per Day')
plt.xlabel('Day')
plt.ylabel('Percentage')
plt.grid()


for p in ax.patches:
    percentage = '{:.1f}%'.format( p.get_width() )
    ax.annotate(percentage, 
                (p.get_width()-0.2, p.get_y() + p.get_height() / 2),  # posisi teks
                ha='left', va='center', 
                xytext=(5, 0), 
                textcoords='offset points') 

plt.show()
```

```python
count_by_day=tips.groupby(['day','sex'])['tip'].count().reset_index().rename(columns={'tip':'count'})
# Menambahkan anotasi dalam bentuk persentase pada setiap batang plot
# Menghitung total jumlah data
count_by_day['total_count'] = count_by_day['count'].sum()

# Menghitung persentase
count_by_day['percentage'] = (count_by_day['count'] / count_by_day['total_count']) * 100

count_by_day
```

```python

ax=sns.barplot(data=count_by_day,x='day',y='count',hue='sex')
plt.title('Count of Customer per Day')
plt.xlabel('Day')
plt.ylabel('Count of Customer')
plt.grid()

# Menambahkan anotasi pada setiap batang plot
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()-1), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points')

plt.show()
```

```python
plt.figure(figsize=(15,8))
ax=sns.barplot(data=count_by_day,x='count',y='day',hue='sex')
plt.title('Count of Customer per Day')
plt.xlabel('Day')
plt.ylabel('Count of Customer')
plt.grid()

# Menambahkan anotasi pada setiap batang plot
for p in ax.patches:
    if p.get_width()>0:
        ax.annotate(format(p.get_width(), '.0f'), 
                    (p.get_width(), p.get_y() + p.get_height() / 2),  # posisi teks
                    ha='left', va='center', 
                    xytext=(5, 0), 
                    textcoords='offset points') 
    else:
        break

plt.show()
```

# Pie Chart

## Matplotlib

```python
count_by_day=tips.groupby(['day'])['tip'].count().reset_index().rename(columns={'tip':'count'})
# Menambahkan anotasi dalam bentuk persentase pada setiap batang plot
# Menghitung total jumlah data
count_by_day['total_count'] = count_by_day['count'].sum()

# Menghitung persentase
count_by_day['percentage'] = (count_by_day['count'] / count_by_day['total_count']) * 100

count_by_day

```

```python
plt.pie(data=count_by_day,x='percentage',labels='day')
plt.title('Customer Proportion per Day')
plt.show()
```

```python

plt.pie(data=count_by_day,x='percentage',labels='day', autopct='%.2f%%'
       )
plt.title('Customer Proportion per Day')
plt.show()
```

```python

plt.pie(data=count_by_day,x='percentage',labels='day', autopct='%1.1f%%'
       ,explode=(0,0.1,0.0,0))
plt.title('Customer Proportion per Day')
plt.show()
```

## Pandas

```python
count_by_day.plot.pie(y='percentage',labels=count_by_day['day'],figsize=(8,8))
plt.title('Customer Proportion per Day')
plt.show()
```
