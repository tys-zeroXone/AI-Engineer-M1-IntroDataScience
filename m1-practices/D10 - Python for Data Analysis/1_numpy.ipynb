hati-hati ketika mengubah isi array

```python
array4 = np.arange(0,11,1)
array4 
```

```python
slice_of_array4 = array4[0:3]
slice_of_array4  
```

```python
# mengubah semua item dalam slice_of_array4 menjadi 88
slice_of_array4[:] = 88
slice_of_array4 
```

```python
array4 # item dalam array4 ikut berubah 
```

Cara yang benar pakai copy()

```python
array4 = np.arange(0,11,1)
array4 
```

```python
slice_of_array4 = array4[0:3].copy() 
slice_of_array4  
```

```python
slice_of_array4[:] = 88
slice_of_array4 
```

```python
array4 # aman, item dalam array4 tidak ikut berubah
```

### Concatenate

```python
arr1 = np.array([[1,2], [3,4], [5,6]])
arr1 
```

```python
arr2 = np.array([[10,20], [30,40]])
arr2 
```

```python
np.concatenate((arr1, arr2), axis=0)
# axis=0 --> menambah baris atau menempelkan ke bawah
```

```python
np.concatenate((arr1, arr2), axis=1)
# axis=1 --> menambah kolom atau menempelkan ke samping 
# error karena jumlah barisnya berbeda
```
