## 1. Create Series

- Series mirip dengan 1D array
- Series bisa kita anggap sebagai 1 kolom dalam sebuah tabel

**membuat series menggunakan list**

```python
list_number = [10,20,30,40]
list_label = ['a','b','c','d']

pd.Series(data=list_number, index=list_label) 
```

**membuat series menggunakan array**

```python
myArray = np.arange(10,41,10)

pd.Series(data=myArray, index=list_label)  
```

**membuat series menggunakan dictionary**

```python
myDict = {'x':11, 'y':12, 'z':13}

pd.Series(data=myDict)
# key akan menjadi label index
# value akan menjadi itemnya
```

```python
myDict = {'x':11, 'y':12, 'z':13.7}
pd.Series(data=myDict)
# kalau 1 item ada yg float, maka semua item berubah jadi float
```

```python
myDict = {'x':11, 'y':'12', 'z':13.7}
pd.Series(data=myDict)
# kalau 1 item ada yg string, maka semua item berubah jadi string (object)
```

## 2. Create DataFrame

**Soal**

## 4. Conditional Indexing

Mencari baris berdasarkan kondisi yg True

```python
df
```

```python
# kondisi
df['name'] == 'Ana De Armas'
```

```python
# conditional indexing
# df[kondisi]

df[df['name'] == 'Ana De Armas']
```

```python
# soal: tampilkan karyawan yg salary-nya lebih dari 7000
df[df['salary']>7000]
```

```python
# menampilkan karyawan yg salary-nya lebih dari 7000 dan gendernya M

# kondisi 1
df['salary']>7000
```

```python
# kondisi 2
df['gender']=='M'
```

```python
# df[(kondisi1) & (kondisi2)]

# &: and --> semua kondisi harus True
# |: or  --> salah satu kondisi harus True

df[(df['salary']>7000) & (df['gender']=='M')]
```
